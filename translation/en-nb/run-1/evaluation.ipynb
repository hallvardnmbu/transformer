```python
prompts = [
    "Universe",
    "A chair.",
    "My book is on the table.",
    "I like sitting at school when working.",
    "By attending the physics-lectures I got sleepy.",
    "Who could avoid death?",
    "Norwegian University of Life Sciences"
]
models = {
    "0 epochs": Translator(),
    "27 epochs": torch.load("./model-27.pth", map_location=torch.device('cpu')),
}
```

# Example translations from the untrained and checkpointed models

```python
for state, model in models.items():
    model.config.device = "cpu"
    
    print(f"Model: {state}")
    for prompt in prompts:
        print(f" From: {prompt}")
        print(f"   To: {model(prompt)}")
        print("------")
    print()
```
